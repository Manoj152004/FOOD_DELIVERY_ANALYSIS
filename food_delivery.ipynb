```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\System -02\Downloads\Food Delivery Cleaned 02.csv")
df.drop_duplicates()
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
df.rename(columns={'delivery_partner_id':'delivery_boy_id'},inplace = True)
df.head(5)
```

```python
df.isnull().sum().sort_values(ascending=False).head(5)
```

```python
for i in ['restaurant_name','cuisine_type','peak_hour','order_status','city']:
    df[i] = df[i].str.title()
```

```python
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],format= 'mixed',dayfirst = True)
```

```python
df['time'] = pd.to_datetime(df['time'],format = '%H:%M').dt.time
```

```python
df.isnull().sum().sort_values(ascending=False).head(5)
```

```python
for col in  ['distance_km','discount_applied_inr','rating']:
    df[col] = df[col].fillna(df[col].median())
```
